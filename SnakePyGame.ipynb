```python
import pygame
from pygame.locals import *
from sys import exit
from random import randint

#posição de spawn da snake
spawn_x = l/2
spawn_y = a/2 

#posicao da maçã
rand_x = randint(10,340)
rand_y = randint(10,340)

#direção da snake
velocidade = 5
x_dir = velocidade
y_dir = 0

vel_clock = 30 #velocidade inicial do clock
dead = False

score = 0

#comprimento da cobra
comprimento = 10

l = 350 #largura da tela
a = 350 #altura da tela

pygame.init()
screen = pygame.display.set_mode((l,a))
pygame.display.set_caption('Snake Game')
clock = pygame.time.Clock()

lstCorpo = []
def add_size(lstCorpo):
    for XeY in lstCorpo:
        pygame.draw.rect(screen,(0,160,0),(XeY[0],XeY[1],10,10))

def restart(): #reinicia os parâmetros iniciais
    global comprimento,spawn_x,spawn_y,rand_x,rand_y,lstCabeca,lstCorpo,dead, score
    comprimento = 10
    spawn_x = l/2
    spawn_y = a/2 
    rand_x = randint(10,340)
    rand_y = randint(10,340)
    lstCabeca = []
    lstCorpo = []
    dead = False
    score = 0
    
while True:
    
    screen.fill((0,0,0)) #cor de fundo
    clock.tick(vel_clock) #velocidade de clock
    
    snake = pygame.draw.rect(screen,(0,160,0),(spawn_x,spawn_y,10,10)) #desenha a snake
    maca = pygame.draw.rect(screen,(255,0,0),(rand_x,rand_y,10,10)) #desenha a maçã
    
    lstCabeca = []
    lstCabeca.append(spawn_x)
    lstCabeca.append(spawn_y)
    lstCorpo.append(lstCabeca)
    
    if lstCorpo.count(lstCabeca) > 1: #se a cobra encostar nela mesma
        
        title = 'DEAD! PRESS SPACE TO RESTART!!!'
        pygame.display.set_caption(title)
        
        dead = True
        while dead:
            #checa eventos
            for event in pygame.event.get():
                if event.type == QUIT:
                    pygame.quit()
                    exit()
                if event.type == KEYDOWN:
                    if event.key == K_SPACE:
                        print('Restarting game')
                        
                        restart()
                        title = 'Snake Game - Score: ' + str(score)
                        pygame.display.set_caption(title)
    
    if len(lstCorpo) > comprimento:
        del lstCorpo[0]
    
    #se a snake sair da tela, a posição é setada para o lado oposto
    if spawn_x > l:
        spawn_x = 0
    if spawn_x < 0:
        spawn_x = l
    if spawn_y < 0:
        spawn_y = a
    if spawn_y > a:
        spawn_y = 0
    
    
    add_size(lstCorpo)
    
    #incrementa a posição da snake
    spawn_x = spawn_x + x_dir
    spawn_y = spawn_y + y_dir
    
    #checa eventos
    for event in pygame.event.get():
        
        #se o usuario fechar a tela
        if event.type == QUIT:
            pygame.quit()
            exit()
        
        #se o usuário pressionar alguma tecla
        if event.type == KEYDOWN:
            if event.key == K_LEFT:
                if x_dir == velocidade: #se a snake estiver se deslocando para a esquerda
                    pass #ignora o comando
                else:
                    x_dir = - velocidade
                    y_dir = 0
                    
            if event.key == K_RIGHT:
                if x_dir == -velocidade: #se a snake estiver se deslocando para a direita
                    pass #ignora o comando
                else:
                    x_dir = velocidade
                    y_dir = 0
            
            if event.key == K_UP:
                if y_dir == velocidade: #se a snake estiver se deslocando para baixo
                    pass #ignora o comando
                else:
                    x_dir = 0
                    y_dir = - velocidade
                
            if event.key == K_DOWN:
                if y_dir == -velocidade: #se a snake estiver se deslocando para cima
                    pass #ignora o comando
                else:
                    x_dir = 0
                    y_dir = velocidade
                    
    #Se acontecer colisão entre os dois retângulos
    if snake.colliderect(maca):
        print('Velocidade:', velocidade)
        print('Comprimento:', comprimento)
        
        score = score + 1
        print('Score: ', str(score))
        
        #respawn da maçã
        rand_x = randint(10,340)
        rand_y = randint(10,340)
        
        #aumenta comprimento da snake
        comprimento = comprimento + 1
        
        #incrementa velocidade de clock
        if vel_clock < 100:
            vel_clock = vel_clock + 1
         
        title = 'Snake Game - Score: ' + str(score)
        pygame.display.set_caption(title)
        
        
    pygame.display.update()
```
